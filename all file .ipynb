# Train News File

```python
traindf=pd.read_csv("train_1.csv")
adf=traindf
adf
```

```python
adf['title'].fillna("",inplace=True)
```

```python
def findsource(title):
    val=title.split("-")[-1].strip()
    val1=str(val).split()
    if len(val1)<5 and ( str(val)=="Breitbart" or val=="The New York Times" ):
        
        return title.split("-")[-1]
    else:
        return np.nan
adf['source']=adf['title'].apply(findsource)
adf.head(20)
```

```python
import re

adf['title']=adf['title'].apply(lambda x: re.sub('- Breitbart','',x))
adf['title']=adf['title'].apply(lambda x: re.sub('- The New York Times','',x))
adf['title'][19]
```

```python
import random

true_ch=['BBC News','Google News','Associated Press','Fox News','Wall Street Journal','NPR',
         'The Washington Post','CNN']
fake_ch=['70news.wordpress','Abcnews.com.co','Infowars','Rilenews','Worldnewsreport','Empirenews',
         'Huzlers','Theonion','Nationalreport']

for i in adf[adf['source'].isna()==True].index:
    if adf.iloc[i]['label']==1:
        adf.at[i,'source']=random.choice(fake_ch)
    elif adf.iloc[i]['label']==0:
        adf.at[i,'source']=random.choice(true_ch)
adf
```

```python
adf.drop('id',axis=1,inplace=True)
adf.drop('author',axis=1,inplace=True)
adf.iloc[:,[0,1,3,2]]
```

# Fake News File

```python
f = pd.read_csv('Fake.csv')
```

```python
ar1=[]
for i in f['title']:
    if '(' in i:
        val= i.split('(')
        ar1.append('('+val[1])

    else:
        ar1.append(np.nan)
print(ar1)
```

```python
f['source']= ar1
```

```python
f=f.replace('(TWEETS)','(TWEET)')
f=f.replace('(GRAPHIC IMAGES)','(IMAGES)')
f=f.replace('(TWEETS/IMAGES)','(TWEET)')
f=f.replace('(TWEET/VIDEO)','(TWEET)')
f=f.replace('(SCREENSHOTS)','(IMAGES)')
f=f.replace('(IMAGE)','(IMAGES)')
f=f.replace('(VIDEO)','(VIDEOS)')

f=f.replace('(TWEET)','Twitter')
f=f.replace('(IMAGES)','Facebook')
f=f.replace('(VIDEOS)','Youtube')
f=f.replace('(AUDIO)','Youtube')
f=f.replace('(DETAILS)','Other')
```

```python
fake_ch=['70news.wordpress','Abcnews.com.co','Infowars','Rilenews','Worldnewsreport','Empirenews',
         'Huzlers','Theonion','Nationalreport']
for i in f[f['source'].isna()==True].index:
     f.at[i,'source']=random.choice(fake_ch)
f.head(30)
```
